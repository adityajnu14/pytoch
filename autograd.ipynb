```python
import torch
print(torch.__version__)
import torch.nn as nn
```

```python
x= torch.tensor(6.0, requires_grad=True)
print(x)
y = x**2
print(y)
y.backward()
x.grad
```

```python
x = torch.tensor(3.0, requires_grad=True)
print(x)
y = x**2
print(y)
z = torch.sin(y)
print(z)
z.backward()
x.grad
```

```python
#autograd for wx+b
x = torch.tensor(6.7)
print(x)
y = torch.tensor(0.0)
print(y)
w = torch.tensor(1.0, requires_grad=True)
print(w)
b= torch.tensor(0.0, requires_grad=True)
print(b)
z = w * x + b
print(z)
y_pred = torch.sigmoid(z)
print(y_pred)
loss = nn.BCELoss()(y_pred, y)
print(loss)
loss.backward()
print(w.grad)
print(b.grad)
```
